```python
import random

nomes_comuns = [
    "Ana", "Bruno", "Carla", "Daniel", "Eduarda", "Felipe", "Gabriela", "Hugo",
    "Isabela", "João", "Larissa", "Marcelo", "Natália", "Otávio", "Paula", "Rafael",
    "Sofia", "Thiago", "Vitória", "Wagner", "Yasmin", "Zeca"
]

sobrenomes_comuns = [
    "Silva", "Santos", "Oliveira", "Souza", "Lima", "Pereira", "Ferreira", "Rodrigues",
    "Almeida", "Gomes", "Martins", "Carvalho", "Ribeiro", "Costa", "Fernandes"
]

clientes = []
for i in range(20):
    nome_completo = f"{random.choice(nomes_comuns)} {random.choice(sobrenomes_comuns)}"
    renda = random.randint(1500, 25000) # Renda entre 1.500 e 25.000
    score_credito = random.randint(300, 950) # Score entre 300 e 950

    cliente = {
        "nome": nome_completo,
        "renda": renda,
        "score_credito": score_credito
    }
    clientes.append(cliente)

# Opcional: para visualizar os clientes gerados
# for cliente in clientes:
#     print(cliente)
```

```python
for cliente in clientes:
    if cliente["score_credito"] >= 600:
        print(cliente["nome"], "- aprovado")
    else:
        print(cliente["nome"], "- em análise manual")
```

```python
class Conta:
    def __init__(self, titular, saldo):
        self.titular = titular
        self.saldo = saldo

    def depositar(self, valor):
        self.saldo += valor

    def sacar(self, valor):
        if self.saldo >= valor:
            self.saldo -= valor
            print(f"Saque de R${valor:.2f} realizado com sucesso. Novo saldo: R${self.saldo:.2f}")
        else:
            print(f"Erro: Saldo insuficiente. Seu saldo atual é R${self.saldo:.2f}")
```

```python
conta_maria = Conta("Maria Oliveira", 1500.00)
conta_maria.depositar(300.00)
```

```python
import pandas as pd
df = pd.DataFrame(clientes)
df.head()
```

```python
clientes_aprovados = df[df["score_credito"] >= 600]

print("Total de aprovados:", len(clientes_aprovados))
clientes_aprovados
```

```python
clientes_qualificados = df[(df['renda'] > 4000) & (df['score_credito'] > 700)]

print("Clientes com renda > 4000 e score_credito > 700:")
print(clientes_qualificados)

# Explicação do operador '&' no Pandas
print("\n--- Explicação do operador '&' ---")
print("No Pandas, o operador '&' (AND bit a bit) é usado para combinar duas ou mais condições booleanas ao filtrar um DataFrame.")
print("Cada condição (por exemplo, `df['renda'] > 4000`) retorna uma Série de valores booleanos (True/False).")
print("O '&' compara os elementos correspondentes dessas Séries booleanas. Para que um cliente seja incluído no resultado final, ele deve satisfazer *ambas* as condições (ou todas as condições conectadas pelo '&').")
print("É importante envolver cada condição entre parênteses, como em `(df['renda'] > 4000)`, para garantir a precedência correta das operações, pois o '&' tem uma precedência maior que os operadores de comparação como '>' ou '<'.")
```
