# Dados fictícios

```
import random

def generate_cpf():
    """Generates a plausible alphanumeric CPF (e.g., XXX.YYY.ZZZ-DD)."""
    # Generate 9 random digits for the main part
    digits = ''.join([str(random.randint(0, 9)) for _ in range(9)])

    # Calculate the first verifier digit
    sum_val = 0
    for i in range(9):
        sum_val += int(digits[i]) * (10 - i)
    v1 = (sum_val * 10) % 11
    if v1 == 10:
        v1 = 0

    # Append first verifier digit
    digits += str(v1)

    # Calculate the second verifier digit
    sum_val = 0
    for i in range(10):
        sum_val += int(digits[i]) * (11 - i)
    v2 = (sum_val * 10) % 11
    if v2 == 10:
        v2 = 0

    # Append second verifier digit
    digits += str(v2)

    # Format as XXX.YYY.ZZZ-DD
    return f"{digits[0:3]}.{digits[3:6]}.{digits[6:9]}-{digits[9:11]}"


first_names = [
    "Ana", "Bruno", "Carla", "Daniel", "Eduarda", "Felipe", "Gabriela",
    "Henrique", "Isabela", "João", "Larissa", "Marcelo", "Natalia", "Otavio",
    "Patricia", "Rafael", "Sabrina", "Thiago", "Ursula", "Victor"
]

last_names = [
    "Silva", "Santos", "Oliveira", "Souza", "Lima", "Pereira", "Ferreira",
    "Almeida", "Costa", "Rodrigues", "Martins", "Carvalho", "Ribeiro", "Gomes",
    "Dias", "Fernandes", "Gonçalves", "Mendes", "Nunes", "Rocha"
]

random.seed(1)
clientes = []

for _ in range(20):
    nome = random.choice(first_names)
    sobrenome = random.choice(last_names)
    # Ensure unique CPF for each customer, though unlikely to collide in small sample
    cpf = generate_cpf()
    renda = round(random.uniform(2000, 15000), 2)  # Renda entre 2000 e 15000
    score_credito = random.randint(300, 900)  # Score de crédito entre 300 e 900

    cliente = {
        "cpf": cpf,
        "nome": nome,
        "sobrenome": sobrenome,
        "renda": renda,
        "score_credito": score_credito
    }
    clientes.append(cliente)

# Display the generated list of clients
for cliente in clientes:
    print(cliente)
```

# Classe

```
# Definição da classe Conta para representar uma conta bancária
class Conta:
    def __init__(self, titular, saldo):
        # Construtor da classe Conta
        # Inicializa a conta com um titular e um saldo inicial
        self.titular = titular
        self.saldo = saldo
        print(f"Conta criada para {self.titular} com saldo inicial de R$ {self.saldo:.2f}.")

    def depositar(self, valor):
        # Método para depositar um valor na conta
        # Adiciona o valor ao saldo atual
        self.saldo += valor

    def sacar(self, valor):
        # Método para sacar um valor da conta
        # Verifica se há saldo suficiente antes de realizar o saque
        if self.saldo >= valor:
            self.saldo -= valor
            print(
                f"Saque de R$ {valor:.2f} realizado com sucesso. "
                f"Novo saldo: R$ {self.saldo:.2f}"
            )
        else:
            print(
                f"Erro: Saldo insuficiente. "
                f"Seu saldo atual é R$ {self.saldo:.2f}"
            )


# Exemplo de uso da classe Conta

# 1. Criação de uma instância da classe Conta
conta_maria = Conta("Maria Oliveira", 1500.00)

# 2. Impressão do saldo antes do depósito
print(f"Saldo antes do depósito: R$ {conta_maria.saldo:.2f}")

# 3. Realização de um depósito
valor_deposito = 300.00
conta_maria.depositar(valor_deposito)

# 4. Impressão do saldo depois do depósito
print(f"Saldo após o depósito de R$ {valor_deposito:.2f}: R$ {conta_maria.saldo:.2f}")
```

```
# Reutilizando a definição da classe Conta do bloco anterior
# (Assumimos que a classe Conta já está definida e executada no ambiente)

# Exemplo de uso da classe Conta com operações de saque

# 1. Criação de uma nova instância da classe Conta para demonstração de saques
print("\n--- Demonstrando operações de Saque ---")
conta_joao = Conta("João Silva", 2000.00)

# 2. Saque dentro do saldo disponível
print(f"Saldo antes do primeiro saque: R$ {conta_joao.saldo:.2f}")
valor_saque_1 = 500.00
print(f"Tentando sacar: R$ {valor_saque_1:.2f}")
conta_joao.sacar(valor_saque_1)
print(f"Saldo após o primeiro saque: R$ {conta_joao.saldo:.2f}")

# 3. Saque além do saldo remanescente
print(f"\nSaldo antes do segundo saque: R$ {conta_joao.saldo:.2f}")
valor_saque_2 = 1800.00 # Este valor é maior que o saldo restante (2000 - 500 = 1500)
print(f"Tentando sacar: R$ {valor_saque_2:.2f}")
conta_joao.sacar(valor_saque_2)
print(f"Saldo final: R$ {conta_joao.saldo:.2f}")
```

## Método __str__

Cria classe.

```
# Definição da classe Conta para representar uma conta bancária
class Conta:
    def __init__(self, titular, saldo):
        # Construtor da classe Conta
        # Inicializa a conta com um titular e um saldo inicial
        self.titular = titular
        self.saldo = saldo
        print(f"Conta criada para {self.titular} com saldo inicial de R$ {self.saldo:.2f}.")

    def __str__(self) -> str:
        return f"Cliente: {self.titular} | Saldo: R$ {self.saldo:.2f}"

    def depositar(self, valor):
        # Método para depositar um valor na conta
        # Adiciona o valor ao saldo atual
        self.saldo += valor

    def sacar(self, valor):
        # Método para sacar um valor da conta
        # Verifica se há saldo suficiente antes de realizar o saque
        if self.saldo >= valor:
            self.saldo -= valor
            print(
                f"Saque de R$ {valor:.2f} realizado com sucesso. "
                f"Novo saldo: R$ {self.saldo:.2f}"
            )
        else:
            print(
                f"Erro: Saldo insuficiente. "
                f"Seu saldo atual é R$ {self.saldo:.2f}"
            )
```

Cria conta [objeto].

```
conta_pedro = Conta("Pedro Soares", 30000)

print(conta_pedro)
```
