# Juros simples

**Operações: Juros Simples**

```
def format_currency_pt_br(value):
    # Esta função formata um número como Real Brasileiro (R$) com separador de milhar e 2 casas decimais.
    # Exemplo: 1000.50 será formatado como R$ 1.000,50
    return f"R$ {value:,.2f}".replace(',', 'X').replace('.', ',').replace('X', '.')

# Definição dos parâmetros para o cálculo de Juros Simples
capital = 50000  # Valor inicial investido ou emprestado
taxa_mensal = .02  # Taxa de juros mensal (2% ao mês)
meses = 6  # Período de tempo em meses

# Cálculo dos Juros Simples (J = C * i * t)
juros = capital * taxa_mensal * meses
print(f"Cálculo dos Juros: J = C * i * t")
print(f"J = {format_currency_pt_br(capital)} * {taxa_mensal} * {meses}")
print(f"Juros: {format_currency_pt_br(juros)}")

print("\n")
# Cálculo do Montante (Montante = Capital + Juros)
montante = capital + juros
print(f"Cálculo do Montante: M = C + J")
print(f"M = {format_currency_pt_br(capital)} + {format_currency_pt_br(juros)}")
print(f"Montante final: {format_currency_pt_br(montante)}")
```

# Validador de empréstimo (if-else)

```
# Variáveis de entrada
renda = 3000  # Renda mensal do solicitante
parcela = 1000  # Valor da parcela mensal do empréstimo
limite_comprometimento = 0.30  # Percentual máximo da renda que pode ser comprometido (30%)

# Cálculo do limite da parcela com base na renda
limite_parcela = limite_comprometimento * renda

# Exibir os valores para o usuário
print(f"Renda mensal: {format_currency_pt_br(renda)}")
print(f"Parcela do empréstimo: {format_currency_pt_br(parcela)}")
print(f"Limite de parcela permitido ({limite_comprometimento:.0%} da renda): {format_currency_pt_br(limite_parcela)}")

# Validação do empréstimo
# Se a parcela for menor ou igual ao limite permitido, o empréstimo é APROVADO.
# Caso contrário, o empréstimo é NEGADO.
if parcela <= limite_parcela:
  print("\nStatus do Empréstimo: APROVADO! 🎉")
  print("A parcela se encaixa na sua capacidade de pagamento.")
else:
  print("\nStatus do Empréstimo: NEGADO. 😞")
  print("A parcela excede o valor máximo permitido para sua renda.")
```
